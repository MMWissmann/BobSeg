```python
#filename = '/Users/micha/Desktop/Uni und so/CBG HiWi/Downloads/MAX_LP320-06-08_Concat-51-100.tif'
filename = '/Users/micha/Desktop/Uni und so/CBG HiWi/Membrane Localization/Cell_Model-0.0.1-3D_Sim-PSF(0.2 0.2 0.6) PXS(0.1 0.1 0.1).tif'
#filename = '/Users/micha/Desktop/Uni und so/CBG HiWi/Membrane Localization/Cell_Model-0.0.1(wave)-3D_Sim-PSF(0.2 0.2 0.6) PXS(0.1 0.1 0.1).tif'
#filename = '/Users/micha/Desktop/Uni und so/CBG HiWi/Membrane Localization/Cell_Model-0.0.1(wave)-3D_Sim-PSF(0.07 0.07 0.60) PXS(0.03 0.03 0.10).tif'

stack = imread(filename)
segchannel = stack
print(segchannel.shape)
sigma = [0.,2.,2.]
segchannel_smooth = gaussian(segchannel,sigma)
```

```python
max_rs = (segchannel[0].shape[1] /2 + 20,
          segchannel[0].shape[1] /2 + 20)

min_rs = (1,1)


data = Data3dRadial( segchannel_smooth, pixelsize=(1.,1.), silent=False )
data.set_seg_params(num_columns=72,K=100,max_delta_k=8)
objects = []
```

```python
cell = data.init_object("cell 1")
objects.append(cell)

data.add_object_at( cell, min_rs, max_rs, frame=8, seed=(67,81), segment_it=True )

data.track(cell, seed_frame=8, target_frames=range(9,39), recenter_iterations=1)
```
